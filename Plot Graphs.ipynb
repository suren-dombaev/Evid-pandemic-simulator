```python
import pickle
import matplotlib.pyplot as plt
from os import path 
```

```python
with open('./experiments/1/checkpoint_2021-05-21 01_00_00.pickle', 'rb') as f:
    model = pickle.load(f)

df = model.datacollector.get_model_vars_dataframe()
df['Not Infected'] = 0
plt.stackplot(range(len(df)),
                    df.loc[:, ['Infected', 'Not Infected', 'Dead', 'Healed']].to_numpy().T,
                    labels=['Infected', 'Not Infected', 'Dead', 'Healed' ],
                    alpha=1,
                    colors=['red', 'whitesmoke', 'gray', 'limegreen'])
plt.ylim([0,10000])
plt.show()
```
